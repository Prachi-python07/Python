```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5]
temp=[28.2,30.4,25.9,32.8,29.7]
```

```python
plt.xlim([0,6])
plt.ylim([0,40])
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("temperature graph")
plt.plot(days,temp,color='m',marker='*',linestyle='-.')
```

```python
company=['motorola','samsung','xiomi']
unitsale=[17600,19200,12700]
```

```python
plt.bar(company,unitsale,color=['c','g','k'])
```

```python
plt.pie(unitsale,labels=company,autopct="%0.2f%%")
```

```python
age=[21,20,21,21,24,23,23,23,21]
plt.hist(age,bins=[20,21,22,23,24,25],histtype='step')
plt.ylabel("count")
plt.xlabel("age")
plt.title("distributuion")
```
