```python
fig,ax=plt.subplots(1,2,figsize=(12,5))
ax[0].scatter(df.sepal_length,df.petal_length)
ax[0].set_xlabel("sepal_length")
ax[0].set_ylabel("petal_length")
ax[0].set_title("scatter plot")
ax[1].scatter(df.sepal_length,df.petal_width)
ax[1].set_xlabel("sepal_length")
ax[1].set_ylabel("petal_width")
ax[1].set_title("scatter plot")
plt.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(15,10))
ax[0,0].scatter(df.sepal_length,df.petal_length)
ax[0,0].set_xlabel("sepal_length")
ax[0,0].set_ylabel("petal_length")
ax[0,0].set_title("scatter plot")
ax[0,1].scatter(df.sepal_length,df.petal_width)
ax[0,1].set_xlabel("sepal_length")
ax[0,1].set_ylabel("petal_width")
ax[0,1].set_title("scatter plot")
ax[1,0].scatter(df.sepal_width,df.petal_length)
ax[1,0].set_xlabel("sepal_width")
ax[1,0].set_ylabel("petal_length")
ax[1,0].set_title("scatter plot")
ax[1,1].scatter(df.sepal_width,df.petal_width)
ax[1,1].set_xlabel("sepal_width")
ax[1,1].set_ylabel("petal_width")
ax[1,1].set_title("scatter plot")
plt.show()
```

```python
# prompt: want to increase the vertical and horizontal space between subplots

fig,ax=plt.subplots(2,2,figsize=(15,10), constrained_layout=False)
fig.subplots_adjust(wspace=0.3,hspace=0.3) #increase spacing between subplots

ax[0,0].scatter(df.sepal_length,df.petal_length,color='g')
ax[0,0].set_xlabel("sepal_length")
ax[0,0].set_ylabel("petal_length")
ax[0,0].set_title("scatter plot")

ax[0,1].scatter(df.sepal_length,df.petal_width,color='r')
ax[0,1].set_xlabel("sepal_length")
ax[0,1].set_ylabel("petal_width")
ax[0,1].set_title("scatter plot")

ax[1,0].scatter(df.sepal_width,df.petal_length)
ax[1,0].set_xlabel("sepal_width")
ax[1,0].set_ylabel("petal_length")
ax[1,0].set_title("scatter plot")

ax[1,1].scatter(df.sepal_width,df.petal_width,color='k')
ax[1,1].set_xlabel("sepal_width")
ax[1,1].set_ylabel("petal_width")
ax[1,1].set_title("scatter plot")

plt.show()
```
